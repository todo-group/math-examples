* $\displaystyle A = \begin{bmatrix} 3 & 2 \\ 1 & 0 \\ 4 & 7 \end{bmatrix}$
* $\displaystyle B = \begin{bmatrix} 1 & -1 & 0 \\ 5 & 1 & -3 \\ 2 & 0 & 7 \end{bmatrix}$
* $\displaystyle C = \begin{bmatrix} 2 & -1 \\ 0 & 5 \end{bmatrix}$
* $\displaystyle D = \begin{bmatrix} -1 \\ 0 \\ 5 \end{bmatrix}$
* $\displaystyle F = \begin{bmatrix} 0 & -1 & 3 \end{bmatrix}$

* 積が定義されるのは、$AC, BA, BD, DF, FB, FD$
* 同じ行列の積を含めてよいなら$BB, CC$も積が定義される

```python
A = numpy.array([[3,2],[1,0],[4,7]]);
B = numpy.array([[1,-1,-1],[5,1,-3],[2,0,7]]);
C = numpy.array([[2,-1], [0,5]]);
D = numpy.array([[-1],[0], [5]]);
F = numpy.array([[0,1,3]]);
print('AC =', A @ C);
print('BA =', B @ A);
print('BD =', B @ D);
print('DF =', D @ F);
print('FB =', F @ B);
print('FD =', F @ D);
print('BB =', B @ B);
print('CC =', C @ C);
```

### 問3-2
$\displaystyle A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$ が $A^2 - 2A -3E=0$を満たすとき、$a+d$、$ad-bc$の値を求めよ

* ケーリーハミルトンの定理より、$A^2 - (a+d) A + (ad-bc)E = 0$が成り立つ。これと $A^2 - 2A - 3E=0$ を見比べて、$a+d=-2, ad-bc=-3$が得られる。確かにこれは一つの解なのだが、実は他にも解がある
* $A^2 - (a+d) A + (ad-bc) = 0$より、$A^2 = (a+d) A - (ad-bc)E$が成り立つ。これを $A^2 - 2A - 3E=0$ に代入すると<br>
  $(a+d) A - (ad-bc)E - 2A - 3E = 0$ ⇒ $(a+d-2) A = (ad-bc+3) E$<br>
* $a+d-2=0$のとき、左辺は$O$となるので、$ad-bc+3=0$ ⇒ 上記の解
* $a+d-2 \ne 0$のとき、両辺を$a+d-2$で割って、$\displaystyle  A =  \frac{ad-bc+3}{a+d-2} E = kE$ ($\displaystyle k = \frac{ad-bc+3}{a+d-2}$とおいた)
  * これを元の式に代入すると、$k^2E - 2kE - 3E =  (k^2-2k-3)E = (k-3)(k+1)E = 0$ ⇒ $k=3,-1$
  * $k=3$のとき ⇒ $A = 3E$ ⇒ $a+d=6, ad-bc=9$
  * $k=-1$のとき ⇒ $A = -E$ ⇒ $a+d=-2, ad-bc=1$
* 以上をまとめて、$(a+d,ad-bc) = (2,-3), (6,9), (-2,1)$
