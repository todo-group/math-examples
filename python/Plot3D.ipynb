# 二変数関数のプロット

 ## ライブラリのインポート

```python
%matplotlib inline
import numpy
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
```

## 一変数関数のプロット

```python
def f(x):
    return x**2;

x = numpy.linspace(-5,5);
pyplot.plot(x, f(x));
```

```python
x = numpy.linspace(-5,5);
pyplot.plot(x, x**2);
```

## 二変数関数のプロット

### データの生成

```python
def f(x, y):
    return x**2 + y**2

x, y = numpy.mgrid[-5:5:0.2, -5:5:0.2];
z = f(x, y);
```

```python
print(x);
print(y);
print(z);
```

### 表面プロット

```python
fig = pyplot.figure();
ax = fig.add_subplot(111, projection='3d');
ax.plot_surface(x, y, z);
```

### ワイヤープロット

```python
fig = pyplot.figure();
ax = fig.add_subplot(111, projection='3d');
ax.plot_wireframe(x, y, z);
```

### 散布図

```python
fig = pyplot.figure();
ax = fig.add_subplot(111, projection='3d');
ax.scatter(x, y, z, s = 1);
```

### 等高線

```python
pyplot.contour(x, y, z);
```

### 色付き等高線

```python
pyplot.contourf(x, y, z);
```

### 縦と横のアスペクト比を1:1に

```python
pyplot.contour(x, y, z);
pyplot.gca().set_aspect('equal');
```

### 複雑な関数のプロット

```python
def f(x, y):
    return 3*(1-x)**2 * numpy.exp(-(x**2) - (y+1)**2) - 10*(x/5 - x**3 - y**5) * numpy.exp(-x**2-y**2) - 1/3 * numpy.exp(-(x+1)**2 - y**2);
fig = pyplot.figure();
ax = fig.add_subplot(111, projection='3d');
ax.plot_surface(x, y, f(x, y));
```
