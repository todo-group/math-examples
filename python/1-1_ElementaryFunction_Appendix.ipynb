## 対数関数

* 指数関数の逆関数
* 年5\%の複利で倍になるまでにかかる年数 $\rightarrow$ $(1.05)^n = 2$ となる $n$
* $\log_{1.05} 2$ と書く (14.2年くらい)
* 一般に $a^y = x$ と満たす $y$ を $\log_a x$ と書く($a$ は 1 でない正の定数)。$a$ を底とする対数関数と呼ぶ
* 対数関数の性質
  * $\log_a 1 = 0$  ($a^0 = 1$ より)
  * $\log_a a = 1$  ($a^1 = a$ より)
  * $\log_a (xy) = \log_a x + \log_a y$
  * $a^{\log_a x} = x$
  * $\log_a (x^p) = p \log_a x$
  * $a^x = b^{x \log_b a}$
  * $\log_a x = (\log_b x) / (\log_b a)$
* 底$a$が10の場合 $\rightarrow$ 常用対数。例: $\log_{10} 1000 = 3$
* 底$a$が2の場合
  * コンピュータサイエンスでよく出てくる
  * 2進数を考えるときに便利
  * 例: $\log_2 1000 \simeq 10$ $(2^10=1024)$
* 底が $e = 2.71828182846\cdots$ (ネイピア数)の場合 $\rightarrow$ 自然対数、単に $\log x$ あるいは $\ln x$ と書く
* 底を $e$ にすると微分が便利

### 底が0.5、2、$e$、10の場合の対数関数の振る舞い

```python
x = numpy.linspace(0.2,15);
pyplot.plot(x, numpy.log(x) / numpy.log(0.1), label='log_0.5');
pyplot.plot(x, numpy.log2(x), label='log_2');
pyplot.plot(x, numpy.log(x), label='ln');
pyplot.plot(x, numpy.log10(x), label='log_10');
pyplot.grid(which="major");
pyplot.legend();
```

### 対数関数の導関数

* 対数関数は指数関数の逆関数<br>
* $y = f(x) = \log x$ の導関数
  $f'(x) = 1 / f'^{-1}(y=f(x)) = 1 / e^y = 1 / e^{\log x} = 1/x$
* 対数微分<br>
  $y=f(x)$が微分可能であるとき、$z = \log y = \log f(x)$ の導関数は<br>
  $\displaystyle \frac{dz}{dx} = \frac{dz}{dy} \frac{dy}{dx} = (1/y) f'(x) = \frac{f'(x)}{f(x)}$
* 底が$e$ではない場合<br>
  $y=a^x$の両辺の対数をとると$\log y = x \log a$<br>
  両辺をxで微分すると $y'/y = log a$<br>
  $y' = (\log a) y = (\log a) a^x$

### 対数関数のテイラー展開 (x=aのまわりでの展開)

* $\log x \simeq \log a + (1/a) (x-a) + \cdots$
* $x=1$ の近傍では $\log x \simeq (x-1) - (x-1)^2/2 + (x-1)^3/3 \cdots$
* $x \rightarrow x+1$ と置き換えると $\log (1+x) \simeq x - x^2/2 + x^3/3 \cdots$
  
### 72の法則

* 複利で利子が$r$\%のとき、2倍になるのに要する年数$n$: $(1+r/100)^n = 2$
* 両辺の自然対数を取ると$n \log(1+r/100) = \log 2$
* $r/100$が1より十分小さい時、テイラー展開の1次の項まで取ると$\log(1+r/100) \simeq r/100$<br>
  → $n (r/100) = \log 2$<br>
  → $n = 100 \log 2 / r = 69.31\cdots/r$
* 69.3の代わりに約数の多い72を使うことが多い

```python
x = numpy.linspace(0, 2* numpy.pi);
pyplot.plot(x, numpy.sin(x), label='sin');
pyplot.plot(x, numpy.cos(x), label='cos');
pyplot.plot(x, numpy.sin(x)**2 + numpy.cos(x)**2, label='sin^2+cos^2');
pyplot.grid(which="major");
pyplot.legend();
```

### 三角関数の微分
* まず、$\displaystyle \lim_{x\rightarrow0} \sin x / x = 1$が成り立つことに注意する

```python
Image('fig-1-1-6.png')
```

* また、$(\cos x -1) / x = ((\cos x - 1)(\cos x + 1)/(x (\cos x+1)) = -\sin^2 x / (x(\cos x+1)) = - \sin x \times (\sin x / x) \times (1/(\cos x+1))$ であるので、$x\rightarrow 0$の極限を考えると $(\cos x -1) \rightarrow 0 \times 1 \times 1/2 = 0$
* 三角関数の微分
  \begin{align*}
    (\sin x)' &= \lim_{h\rightarrow0} (\sin(x+h) - \sin x) / h \\
    &= \lim_{h\rightarrow0} (\sin x \cos h + \cos x \sin h - \sin x) / h \\
    &= \lim_{h\rightarrow0} \sin x (\cos h - 1)/h + \cos x \sin h / h \\
    &= \cos x \\
    (\cos x)' &= \lim_{h\rightarrow0} (\cos(x+h) - \cos x) / h \\
    &= \lim_{h\rightarrow0} (\cos x \cos h - \sin x \sin h - \cos x) / h \\
    &= \lim_{h\rightarrow0} \cos x (\cos h - 1)/h - \sin x \sin h / h \\
    &= -\sin x \\
    (\tan x)' &= (\sin x / \cos x)' = ((\sin x)' \cos x - \sin x (\cos x)') / \cos^2 x = 1 / \cos^2 x = 1+\tan^2 x
  \end{align*}
