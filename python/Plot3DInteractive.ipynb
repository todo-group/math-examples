# 回せる3Dプロット

 ## ライブラリのインポート

```python
import numpy
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
```

## 二変数関数のプロット

### データの生成

```python
def f(x, y):
    return x**2 + y**2
x, y = numpy.mgrid[-5:5:0.2, -5:5:0.2];
z = f(x, y);
```

### 表面プロット

```python
%matplotlib notebook
fig = pyplot.figure();
ax = fig.add_subplot(111, projection='3d');
ax.plot_surface(x, y, z);
```

### ワイヤープロット

```python
fig = pyplot.figure();
ax = fig.add_subplot(111, projection='3d');
ax.plot_wireframe(x, y, z);
pyplot.show();
```

### 等高線

### 縦と横のアスペクト比を1:1に

```python
%matplotlib inline
pyplot.contour(x, y, z);
pyplot.gca().set_aspect('equal');
```
